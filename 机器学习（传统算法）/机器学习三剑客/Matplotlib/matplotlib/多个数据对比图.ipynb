```python
from matplotlib import pyplot as plt
a=[1,0,1,2,4,3,2,3,4,4,5,6,5,4,3,3,1,1,1,1]
b=[0,2,1,2,3,4,5,3,3,2,3,0,1,2,3,1,0,0,0,0]
x=range(11,31)
plt.figure(figsize=(20,8),dpi=80)
#label添加标签，color配颜色，linestyle更改风格，linewidth调整粗细
plt.plot(x,a,label='I',color='purple',linestyle='--')
plt.plot(x,b,label='friend',color='black',linestyle=':')
yo_tick_labels=['{}years'.format(i) for i in x]
plt.xticks(x, yo_tick_labels)
plt.yticks(range(0,10))
plt.grid(alpha=0.4)#设置网格,alpha网格透明度
plt.legend(loc=9)#图例添加需要在plot和添加legend,loc=调整图例位置。
plt.show()

```
