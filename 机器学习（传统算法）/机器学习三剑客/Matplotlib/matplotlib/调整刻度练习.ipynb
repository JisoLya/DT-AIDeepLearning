```python
from matplotlib import pyplot as plt
import random
x=range(0,120)
y=[random.randint(20,35) for i in range(120)]
plt.figure(figsize=(20,8),dpi=80) #尺寸调整需要在制表前
plt.plot(x,y)
#调整x的刻度
x1=list(x)
_tick_labels=['10h{}min'.format(i) for i in range(60)]
_tick_labels+=['11h{}min'.format(i) for i in range(60)]
plt.xticks(x1[::5],_tick_labels[::5],rotation=45)
#添加描述信息
plt.xlabel("time")
plt.ylabel('temp')
plt.title('11-12 temp')
plt.show()

```
