```python
#创建数组#
import numpy as np

t1= np.array([1,2,3])
print(t1)
print(type(t1))

t2= np.array(range(10))
print(t2)

t3=np.arange(10)
print(t3)
# np.array(range(x))==np.arange(10)#

t4=np.arange(4,10,2)
print(t4)
#------------------------------------------------------#

#查询数据类型：.dtype#
print(t4.dtype)

#改变数组类型#
t5=np.array(range(4,10,2), dtype="i1")
print(t5,t5.dtype)
t7=t5.astype('int64')
print(t7,t7.dtype)

#numpy中的bool类型
t6=np.array([1,1,0,0,1,0],dtype=bool)
print(t6,t6.dtype)
import random

#numpy中的小数
t8=np.array([random.random()for i in range (10)])
print(t8,t8.dtype)

t9=np.round(t8,2)#np.round的使用方法：np.round(数组，保留几位小数)#
print(t9)
```
